```python
df = df.drop(['@context' , '@type' , 'isPartOf' , 'isAccessibleForFree' , 'image' , 'thumbnailUrl' , 'isBasedOn' ] , axis = 1)
```

```python
df.head()
```

```python
df['author'] = df['author'].map(lambda t: t[0]['name'])
```

```python
df['aggregateRating'].iloc[0]
```

```python
df['RatingValue'] = df['aggregateRating'].map(lambda t: t['ratingValue'] , na_action = 'ignore')
df['RatingCount'] = df['aggregateRating'].map(lambda t: t['ratingCount'] , na_action = 'ignore')
```
