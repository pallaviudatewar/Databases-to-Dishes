```python
from langchain.llms import CTransformers

llm = CTransformers(model="marella/gpt-2-ggml")
```

```python
print(llm("Generate a recipe using the folowing ingredients ; Chicken, Ginger and Garlic"))
```

```python
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler

llm = CTransformers(
    model="marella/gpt-2-ggml", callbacks=[StreamingStdOutCallbackHandler()]
)

response = llm("Generate a recipe using the folowing ingredients ; Chicken, Ginger and Garlic")
```

```python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

template = """Question: {question}

Answer:"""

prompt = PromptTemplate(template=template, input_variables=["question"])

llm_chain = LLMChain(prompt=prompt, llm=llm)

response = llm_chain.run("Generate a recipe using the folowing ingredients ; Chicken, Ginger and Garlic")
```
